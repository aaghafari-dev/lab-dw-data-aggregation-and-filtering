```python
import numpy as np
import pandas as pd
url = 'https://raw.githubusercontent.com/data-bootcamp-v4/data/main/marketing_customer_analysis.csv'
df = pd.read_csv(url)  
df.head(3)
```

```python
# 1. Create a new DataFrame that only includes customers who:
# - have a **low total_claim_amount** (e.g., below $1,000),
# - have a response "Yes" to the last marketing campaign.
df2 = df[(df['Total Claim Amount'] < 1000) & (df['Response'] == 'Yes')]
df2.head(3)
```

2. Using the original Dataframe, analyze:
   - the average `monthly_premium` and/or customer lifetime value by `policy_type` and `gender` for customers who responded "Yes", and
   - compare these insights to `total_claim_amount` patterns, and discuss which segments appear most profitable or low-risk for the company.

```python
# 2. Using the original Dataframe, analyze:
#   - the average `monthly_premium` and/or customer lifetime value by `policy_type` and `gender` for customers who responded "Yes", and
#   - compare these insights to `total_claim_amount` patterns, and discuss which segments appear most profitable or low-risk for the company.
df_yes = df[df['Response'] == 'Yes']
pivot_analysis = df_yes.pivot_table(index=['Policy Type', 'Gender'], values=['Customer Lifetime Value', 'Monthly Premium Auto', 'Total Claim Amount'], aggfunc='mean')
pivot_analysis
```

3. Analyze the total number of customers who have policies in each state, and then filter the results to only include states where there are more than 500 customers.

```python
#3. Analyze the total number of customers who have policies in each state, and then filter the results to only include states where there are more than 500 customers.
df_totalNumber = df.groupby('State').size().reset_index(name='num_customers').sort_values(by='num_customers', ascending=False)

# filter states with more than 500 customers
df_totalNumber_filtered = df_totalNumber[df_totalNumber['num_customers'] > 500]

# display the filtered result
df_totalNumber_filtered
```

4. Find the maximum, minimum, and median customer lifetime value by education level and gender. Write your conclusions.

```python
#4. Find the maximum, minimum, and median customer lifetime value by education level and gender. Write your conclusions.
df_max = df.pivot_table(index=['Education', 'Gender'], values=['Customer Lifetime Value'], aggfunc='max')
df_max
```

```python
df_min = df.pivot_table(index=['Education', 'Gender'], values=['Customer Lifetime Value'], aggfunc='min')
df_min
```

```python
df_median = df.pivot_table(index=['Education', 'Gender'], values=['Customer Lifetime Value'], aggfunc='median')
df_median
```
